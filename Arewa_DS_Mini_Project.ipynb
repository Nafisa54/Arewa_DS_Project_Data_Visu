```python
from random import choice
import matplotlib.pyplot as plt
class RandomWalk():
    '''An attempt to model the random motion of a particle within a box using the RandomWalk class'''
    def __init__(self,num_point=50000):
        '''Initialize the attributes of a walk'''
        self.num_point = num_point

    # All walk start at the origin (0,0)
        self.x_values = [0]
        self.y_values = [0]

    # Create a method for the fill work
    def fill_work(self):
        '''Function that returns the x and y values'''

    # Determine the direction and distance for each step
        while len(self.x_values) < self.num_point: # Keep taking steps until the walk reaches the desired length
            x_direction = choice([1,-1]) # Decide which direction to go in that direction
            x_distance = choice([0,1,2,3,4,5,6,7,8]) # Decide how far to go in that direction
            x_step = x_direction * x_distance # Calculate the next x value

            y_direction = choice([1,-1])
            y_distance = choice([0,1,2,3,4,5,6,7,8])
            y_step = y_direction * y_distance

            # Reject moves that go nowhere
            if x_step == 0 and y_step == 0:
                continue

            # Calculate the next x and y values
            next_x = self.x_values[-1] + x_step
            next_y = self.y_values[-1] + y_step

            self.x_values.append(next_x)
            self.y_values.append(next_y)

# Keep working as long as the condition remain true 
while True:
    RandWalk = RandomWalk()
    RandWalk.fill_work()
    num_point_list = list(range(RandWalk.num_point))
    plt.scatter(RandWalk.x_values, RandWalk.y_values,c=num_point_list,cmap=plt.cm.Blues,edgecolors='none', s=20)

# Determine the start point and end point of the walk
    plt.scatter(RandWalk.x_values[0],RandWalk.y_values[0],c='Red',edgecolors='none',s=200)
    plt.scatter(RandWalk.x_values[-1],RandWalk.y_values[-1],c='Green',edgecolors='none',s=200)
    plt.show()
    user_request = input('To continue running the program, enter (y/n)')
    if user_request == 'n':
        break
    else:
        continue
```
